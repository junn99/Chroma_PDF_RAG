# Document & Loader

## Document?

Langchain의 기본 문서 객체로써, 거의 대부분의 함수,클래스들에 이 객체를 값으로 넣어줘야 함!

속성
- page_content(str) : 문서의 내용
- metadata(dict) : 문서의 메타데이터

```python
from langchain_core.documents import Document

document = Document(page_content="이것은 랭체인 다큐먼트의 내용입니다.")
```

```python
document
```

```python
# Dcoument의 속성 값
document.__dict__
```

```python
# metadata 추가
document = Document(page_content="이것은 랭체인 다큐먼트의 내용입니다.", metadata={"source":"jun"})
document.__dict__
```

## Loader

### load() & load_and_split()
- load() : 문서를 List[Document]로 반환
- load_and_split() : load와 split를 한 번에 해줌
    - 사전에 splitter 정의 필요

```python
# data paht
data_path = "/home/jun/my_project/langchain_tutorial/Langcahin_Tutoraial/data/약관-48-51.pdf"
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(data_path)

# 문서 로드 : 페이지 단위
docs = loader.load()

print(docs)
# 로드된 문서 수
print(len(docs))
```

```python
print(docs[1])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

loader = PyPDFLoader(data_path)

# 문서 분할
split_docs = loader.load_and_split(text_splitter=text_splitter)

# 로드된 문서 수 확인
print(len(split_docs))

# 첫번째 문서 확인
print(split_docs[1])
```
